```python
def get_outlier_rows_index(column_name, df):
    Q1 = np.percentile(df[f"{column_name}"], 25)
    Q3 = np.percentile(df[f"{column_name}"], 75)
    IQR = Q3 - Q1
    return df[(df[f"{column_name}"] < Q1 - 1.5*IQR) | (df[f"{column_name}"] > Q3 + 1.5*IQR )].index
```

```python
user_score_outlier_rows = get_outlier_rows_index("user_score", df)
meta_score_outlier_rows = get_outlier_rows_index("meta_score", df)
```

```python
print("user_score_outlier: ",len(user_score_outlier_rows))
print("meta_score_outlier: ",len(meta_score_outlier_rows))
```

### Conclusions:

A helper function for displaying pie graphs:

```python
## if receiving 'top_num' then will display until the number at 'top_num' ##
def make_pie_plot(df, on_column, title, ax, top_num=0):
    vals = df[f"{on_column}"].value_counts().iloc[:top_num].copy()
    others_vals = df[f"{on_column}"].value_counts().iloc[top_num:].sum()
    
    new_df = vals.copy()
    new_df["others"] = others_vals
    if(not top_num):
        new_df = df[f"{on_column}"].value_counts().copy()
    
    mylabels = list(new_df.index)
    ax.set_title(title)
    ax.pie(new_df, autopct='%1.1f%%', labels = mylabels)
    ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title=on_column.title())
```

## Platform:

```python
platform_crosstab = pd.crosstab(df["platform"], df["successful"])
platform_crosstab_sorted = platform_crosstab.sort_values(by=1, ascending=False).copy()

fig = plt.figure(figsize=(10, 3))
ax = plt.axes()
platform_crosstab_sorted.iloc[:15].plot(kind="bar", ax=ax)
ax.set_ylabel("Count")
ax.set_xlabel("Platform")
ax.legend(title="Successful")
plt.show()
```

### Exclusive Games:

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))


df_exclusive = df[df["exclusive_game"] == 1].copy()
df_exclusive_successful = df_exclusive[df_exclusive["successful"] == 1].copy()

make_pie_plot(df_exclusive, "platform", "Exclusive games in general", axes[0])
make_pie_plot(df_exclusive_successful, "platform", "Exclusive successful games", ax=axes[1])
plt.show()
```
