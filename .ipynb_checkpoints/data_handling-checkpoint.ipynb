```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# בדיקט הנתונים בקובץ של המשחקים הטובים
file_name = "games_data/top_rated_games.csv"
games_data = pd.read_csv(file_name, header=0, sep=',')
games_data.head()
```

```python
games_data.info()
```

```python
# ניתן ליראות כי יש בדירוג של המשתמשים נתון בשם tdb
# המציין שעוד לא הוחלט על דירוג
games_data[games_data["user_score"] == "tbd"]
```

```python
# replace all 'tbd' values with nan values
games_data.replace("tbd", np.nan, inplace=True)

# and now ther is no 'tbd's
games_data[games_data["user_score"] == "tbd"]
```

```python
# הנתונים בעלי המיספרים הגדולים עם פסיקים במיספרים
# ולכן עמרה פשוטה לסוג אינט לא תעזור פה
games_data["user_positive_review"][0]
```

```python
# עמרה של כל הנתוני review
# בעלי המיספרים הגבועים
colums = ["user_positive_review", "user_negative_review", "user_mixed_review"]
for column in colums:
    nums = []
    for string in games_data[column]:
        if(type(string) != float):
            nums.append(int(string.replace(",", "")))
        else:
            nums.append(np.nan)
    games_data[column] = nums
    #games_data[column] = games_data[column].astype('int')

games_data.iloc[:,11:14]
```

```python
games_data.head()
```

```python
# מילוי המקומות עם nan
# בממוצע של העמודה
games_data["user_positive_review"].fillna(games_data["user_positive_review"].mean(), inplace=True)
games_data["user_negative_review"].fillna(games_data["user_negative_review"].mean(), inplace=True)
games_data["user_mixed_review"].fillna(games_data["user_mixed_review"].mean(), inplace=True)
```

```python
games_data[games_data["user_negative_review"].isna()]
```

```python
# change the 'user_score' so it will be an int num
# like the 'meta_scroe'
games_data['user_score'] = games_data['user_score'].astype('float')

user_df = games_data.iloc[:,6:7].copy()
user_df = user_df[user_df["user_score"].notnull()] * 10


user_df.head()
games_data['user_score'] = user_df["user_score"]
games_data['user_score'].fillna(games_data["user_score"].mean(), inplace=True)
games_data['user_score'] = games_data['user_score'].astype('int64')
games_data.head()
```

```python
games_data.info()
```

```python
games_data.iloc[:,11:14]
```

```python
# making a new normalize column of 'user_positive_review'
user_review = games_data.iloc[:,11:14].copy()
user_review_sum = user_review["user_positive_review"] + user_review["user_negative_review"] + user_review["user_mixed_review"]
user_review["user_positive_normalize"] = user_review["user_positive_review"] / user_review_sum

games_data["user_positive_normelize"] = user_review["user_positive_normalize"]
games_data.head()
```

```python
# making a new normalize column of 'critic_positive_review'
critic_review = games_data.iloc[:,-4:-1].copy()
critic_review_sum = critic_review["critic_positive_review"] + critic_review["critic_mixed_review"] + critic_review["critic_negative_review"]
critic_review["critic_positive_normalize"] = critic_review["critic_positive_review"] / critic_review_sum

games_data["critic_positive_normelize"] = critic_review["critic_positive_normalize"]
games_data.head()
```
